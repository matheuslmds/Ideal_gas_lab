# Movimento das partículas

Consideremos uma partícula movendo-se com velocidade constante.

Em um intervalo de tempo Δt, sua posição será atualizada segundo:

x(novo) = x(atual) + vx.Δt \
y(novo) = y(atual) + vy.Δt

Como a simulação é discreta, o movimento contínuo será aproximado por sucessivas atualizações temporais.

# Colisões

No modelo atual, as partículas se movem livremente dentro de uma região retangular que representa o recipiente do gás. Para garantir que o sistema permaneça confinado, é necessário implementar colisões entre as partículas e as paredes da caixa.

Fisicamente, assumimos que essas colisões são perfeitamente elásticas, ou seja, não há perda de energia cinética durante o impacto. A única mudança ocorre na direção do movimento da partícula.

Quando uma partícula colide com uma parede vertical (esquerda ou direita), sua componente horizontal da velocidade é invertida:

vx = -vx

De forma análoga, ao colidir com uma parede horizontal (superior ou inferior), ocorre a inversão da componente vertical da velocidade:

vy = -vy


Além disso, é importante garantir que a partícula não permaneça fora dos limites da caixa após a colisão. Por isso, sua posição é corrigida para o ponto mais próximo válido dentro do recipiente antes da inversão da velocidade. Essa correção evita problemas numéricos comuns em simulações discretas, como “penetração” da partícula na parede.

Esse modelo simplificado permite reproduzir, de forma computacional, o comportamento essencial de partículas confinadas, preservando a energia total do sistema e servindo como base para a construção de um gás ideal bidimensional.

```python
import pygame as pyg

pyg.init() #inicializa o pygame
tela = pyg.display.set_mode((800,600)) #Cria a janela e define o seu tamanho

#Título da janela
pyg.display.set_caption("Ideal Gas Lab")

#loop principal
rodando = True

#Definindo a variável clock como
clock = pyg.time.Clock()
    
#Criação do recipiente do gás
caixa_x = 25
caixa_y = 25

caixa_largura = 550
caixa_altura = 550

#Definindo a posição da partícula, sua velocidade e o seu raio
x = caixa_x + caixa_largura/2
y = caixa_y + caixa_altura/2

vx = 2
vy = 1

raio = 10


#Loop em que a simulação ocorrerá
while rodando:
    #Captura eventos
    for evento in pyg.event.get():
        if evento.type == pyg.QUIT:
            rodando = False
    #Deixa o fundo cinza
    tela.fill((230, 230, 230))

    #Desenho do recipiente
    pyg.draw.rect(
        tela,
        (20, 20, 20),
        (
            caixa_x,
            caixa_y,
            caixa_largura,
            caixa_altura
        ),
        width=2
    )

    
    #Cria a partícula
    pyg.draw.circle(
    tela,
    (0, 0, 255),
    (int(x), int(y)),
    raio)

    #Define a sua velocidade em x e y
    x += vx
    y += vy


    #Criação das colisões para o teto e para o chão (lados direito e esquerdo)

    #Parede esquerda
    if x <= caixa_x + raio:
        x = caixa_x + raio
        vx = -vx
    #parede direita
    if x >= caixa_x + caixa_largura - raio:
        x = caixa_x + caixa_largura - raio
        vx = -vx
    #Teto superior
    if y <= caixa_y + raio:
        y = caixa_y + raio
        vy = -vy
    #Teto inferior
    if y >= caixa_y + caixa_altura - raio:
        y = caixa_y + caixa_altura - raio
        vy = -vy
    
    #Limitando a atualização da tela
    #60 iterações por segundo (60 fps)
    clock.tick(60)
    
    #Atualiza a tela
    pyg.display.flip()







pyg.quit()
```
