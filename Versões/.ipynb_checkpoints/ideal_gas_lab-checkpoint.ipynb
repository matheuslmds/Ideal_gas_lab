# Ideal gas lab 

Simulação computacional de um gás ideal utilizando Python e Pygame.

O projeto foi desenvolvido com o objetivo de explorar conceitos da Teoria Cinética dos Gases, permitindo visualizar o movimento das partículas, colisões elásticas, temperatura e distribuição de velocidades em um sistema idealizado.


## Base teórica
### Distribuição de Maxwell-Boltzmann

Em um gás ideal em equilíbrio térmico, as moléculas não possuem todas a mesma velocidade. Em vez disso, as velocidades seguem uma distribuição estatística conhecida como **Distribuição de Maxwell-Boltzmann**, que descreve a probabilidade de uma molécula apresentar determinada velocidade.

De forma qualitativa, a maioria das partículas possui velocidades próximas de um valor mais provável, enquanto poucas partículas apresentam velocidades muito baixas ou muito altas.

Nesta simulação, a distribuição de velocidades é representada por um histograma, construído a partir das velocidades instantâneas das partículas. Embora o modelo utilize um sistema bidimensional simplificado, é possível observar o surgimento de uma distribuição característica semelhante à prevista pela teoria cinética dos gases.

Assim, o histograma permite relacionar o comportamento microscópico das partículas às previsões estatísticas da Física Estatística.

### Temperatura do sistema

Na Teoria Cinética dos Gases, a temperatura está relacionada à energia cinética média das partículas do sistema. Para uma partícula de massa (m), a energia cinética é dada por:

[
E_c = \frac{1}{2}mv^2
]

Em um gás ideal, a temperatura é proporcional à energia cinética média das partículas:

[
T \propto \langle E_c \rangle
]

Como a simulação utiliza partículas de massa unitária e não realiza conversões para unidades físicas reais, a temperatura exibida não corresponde a uma temperatura em Kelvin. Em vez disso, é utilizada uma **temperatura arbitrária**, denotada por (T^*), expressa em **unidades arbitrárias (u.a.)**.

O valor de (T^*) é calculado a partir da média dos quadrados das velocidades das partículas:

[
T^* \propto \frac{1}{N}\sum_{i=1}^{N}(v_{x,i}^2 + v_{y,i}^2)
]

Dessa forma, o aumento da velocidade média das partículas resulta em um aumento da temperatura exibida, preservando a proporcionalidade prevista pela Teoria Cinética dos Gases.

## Criação e movimentação das partículas

As funções desta seção são responsáveis pela geração das partículas e pela atualização contínua de suas posições ao longo da simulação.

```python
def cria_particulas(particulas, num_particulas):
    """Função que cria as particulas, recebendo uma lista vazia e um número de particulas para a simulação e retorna a lista com "n" particulas e as suas propriedades (velocidade e posição)."""

    for _ in range(num_particulas):
        
        #angulo de direção alearorio para cada partícula
        angulo = rd.uniform(0, 2 * mt.pi)

        #velocidade aleatoria para cada partícula
        velocidade = rd.uniform(1, 6) 
        
        #direções aleatorias
        x = rd.randint(
        caixa_x + raio,
        caixa_x + caixa_largura - raio
        )
        y = rd.randint(
        caixa_y + raio,
        caixa_y + caixa_altura - raio
        )
        
        #velocidades iguais, porém em direções aleatórias
        vx = velocidade*mt.cos(angulo)
        vy = velocidade*mt.sin(angulo)
        
        #conjunto dentro de uma lista
        particulas.append(
        {
            "x": x,
            "y": y,
            "vx": vx,
            "vy": vy
        }
        )
    
    return particulas
```

```python
def mover_particulas(particulas):
    """Função que soma a velocidade de cada partícula à sua posição, dando a ideia de movimento."""
    #Atualiza a posição de cada particula (ideia de movimento)
    for particula in particulas:
        particula["x"] += particula["vx"]
        particula["y"] += particula["vy"]
```

## Colisões

As colisões são tratadas como perfeitamente elásticas, isto é, colisões em que o choque entre os corpos conserva tanto a energia cinética total quanto a quantidade de movimento do sistema. Isso significa que, após a colisão, os objetos continuam se movendo sem perda de energia para calor, som ou deformações permanentes, como se “ricocheteassem” um no outro.

Nesse caso, ocorre uma colisão perfeitamente elástica entre duas massas iguais (m1=m2). Portanto, o resultado principal será que os corpos **trocam suas velocidades**.

\[
v1' = v2
\]
\[
v2' = v1
\]


São consideradas tanto as colisões entre partículas quanto as colisões com as paredes do recipiente.

```python
def colisoes_parede(particulas):
    """Função que recebe a lista das particulas e inverte o sentido das suas velocidades quando elas colidem com a parede (colisões elasticas, retornando as velocidades invertidas das particulas."""
    for particula in particulas:
        #Criação das colisões para o teto e para o chão (lados direito e esquerdo)
    
        #Parede esquerda
        if particula["x"] <= caixa_x + raio:

            particula["x"] = caixa_x + raio

            particula["vx"] *= -1
        #parede direita
        if particula["x"] >= caixa_x + caixa_largura - raio:
            particula["x"] = caixa_x + caixa_largura - raio
            particula["vx"] *= -1
    
        #Teto superior
        if particula["y"] <= caixa_y + raio:
            particula["y"] = caixa_y + raio
            particula["vy"] *= -1
    
        #Teto inferior
        if particula["y"] >= caixa_y + caixa_altura - raio:
            particula["y"] = caixa_y + caixa_altura - raio
            particula["vy"] *= -1
```

```python
def colisoes_particulas(particulas, raio):
    """Função que recebe a lista com as informações das partículas e detecta as colisões entre elas. Nesse código, as suas velocidades são invertidas (colisões elasticas) e um vetor de sobreposição normalizado separa elas imperceptívelmente para que uma nova colisão seja detectada (ler o markdown ao final do código)."""
    
    #Percorrer as partículas da lista e comparar as suas posições para as colisões
    for i in range(len(particulas)):
        #i+1, pois irá partir da próxima posição para vasculhar
        for j in range(i+1, len(particulas)):
            p1 = particulas[i]
            p2 = particulas[j]

            #distancia entre as partículas
            dx = p2["x"] - p1["x"]
            dy = p2["y"] - p1["y"]
            distancia = mt.sqrt((dx**2) + (dy**2))

            #caso elas se encostem (colisão), as velocidades são trocadas (colisão elastica)
            if distancia <= 2*raio and distancia > 0:
                p1["vx"], p2["vx"] = p2["vx"], p1["vx"]
                p1["vy"], p2["vy"] = p2["vy"], p1["vy"]

                #Separar minimamente as particulas para que colisões não sejam infinitas
                
                #normaliza os vetores de distancia entre as particulas
                nx = dx/distancia
                ny = dy/distancia

                #Corrige a sobreposição das partículas, ou seja, move cada partícula minimamente
                sobreposicao = 2*raio - distancia

                #Atualiza a posição das partículas, movendo-as minimamente para longe de si para que outra colisão não seja detectada
                p1["x"] -= nx * sobreposicao / 2
                p1["y"] -= ny * sobreposicao / 2
                p2["x"] += nx * sobreposicao / 2
                p2["y"] += ny * sobreposicao / 2
```

```python
def desenha_particulas(particulas, raio, tela):
    """Função que percorre as particulas e desenha as particulas na tela, recebendo a lista das particulas, o valor do seu raio e a tela da simulação."""

    #desenha as partículas na tela
    for particula in particulas:
        
        #Define a rapidez da partícula
        v = (particula["vx"]**2 + particula["vy"]**2)**0.5
        
        #Define a cor da partícula de acordo com a sua velocidade
        if v < 2:
            cor = (0, 0, 255)

        elif v < 4:
            cor = (0, 255, 0)

        else:
            cor = (255, 0, 0)
            
        #desenha a bolinha
        pyg.draw.circle(
            tela,
            cor,
            (
                int(particula["x"]),
                int(particula["y"])
            ),
            raio
        ) 
```

```python
def desenha_caixa(tela, caixa_x, caixa_y, caixa_largura, caixa_altura):
    """Função que recebe as váriaveis das dimensões da caixa e retorna um desenho na tela da caixa com as dimensões proporcionadas."""

    #Desenho do recipiente
    pyg.draw.rect(
        tela,
        (20, 20, 20),
        (
            caixa_x,
            caixa_y,
            caixa_largura,
            caixa_altura
        ),
        width=2
    )
    
```

```python
def texto_particulas(tela, fonte, particulas, max_particulas):
    """Função que recebe a fonte escolhida e a lista de partículas, retornando a escrita do número de particulas na tela."""
    
    #Mostra o número de partículas na tela
    texto = fonte.render(
    f"N = {len(particulas)} / {max_particulas}",
    True,
    (0, 0, 0)
    )
    tela.blit(texto, (620, 50))
```

```python
def controles(fonte, tela, x, y):

    textos = [
        "UP - Aquecer",
        "DOWN - Resfriar",
        "LEFT - Remove partículas",
        "RIGHT - Adiciona partículas"
    ]

    for i, texto in enumerate(textos):

        superficie = fonte.render(texto, True, (0, 0, 0))
        tela.blit(superficie, (x, y + i * 30))
```

```python
def trata_evento(particulas, rodando):
    """ Processa os eventos do Pygame durante a simulação, permitindo encerrar o programa e controlar propriedades do sistema. A tecla UP aumenta em 10% a velocidade das partículas, a tecla DOWN reduz em 10%, a tecla RIGHT adiciona 5 partículas ao sistema e a tecla LEFT remove 5 partículas. Retorna o estado atualizado da simulação. """
    
    #Captura eventos
    for evento in pyg.event.get():
        
        #Se o evento for sair do aplicativo, encerra-se o while e,consequentemente, a simulação
        if evento.type == pyg.QUIT:
            rodando = False

        #Caso a tecla para baixo seja pressionada, diminui-se a velocidade
        if evento.type == pyg.KEYDOWN:
            if evento.key == pyg.K_DOWN:
                for particula in particulas:
                    #Utiliza um valor levemente aleatório para resfriar as partículas
                    valor_resfria = rd.uniform(0.85, 0.95)

                    particula["vx"] *= valor_resfria
                    particula["vy"] *= valor_resfria

                    #limita a diminuição que o usuario pode fazer da velocidade
                    limitar_velocidade(particula)
            
        #Caso a tecla para cima seja pressionada, aumenta-se a velocidade
            if evento.key == pyg.K_UP:
                for particula in particulas:
                    #Utiliza um valor levemente aleatório para aquecer as partículas
                    valor_aquece = rd.uniform(1.05, 1.15)
                    
                    particula["vx"] *= valor_aquece
                    particula["vy"] *= valor_aquece
        

                    #limita o aumento que o usuario pode fazer da velocidade
                    limitar_velocidade(particula)
                    
        #quando a tecla de seta para a direita for pressionada, adiciona-se 5 partículas
            if evento.key == pyg.K_RIGHT:
                adicionar_particulas(particulas, 5)
                
        #quando a tecla de seta para a direita for pressionada, são removidas 5 partículas
            if evento.key == pyg.K_LEFT:
                remover_particulas(particulas, 5)
    
    return rodando
```

```python
def temp(particulas, fonte, tela):
    """ Calcula a temperatura média do sistema a partir das velocidades das partículas e exibe o valor na tela como texto renderizado. """
    
    #Criação da variavel temperatura
    temperatura = 0
    #Soma das velocidades ao quadrado de todas as partículas
    for particula in particulas:
        vx = particula["vx"]
        vy = particula["vy"]
        temperatura += vy**2 + vx**2

    #Calculo da energia cinetica média das particulas, isto é, a temperatura
    temperatura = temperatura/len(particulas)

    #Arredondando o seu valor para duas casas decimais
    temperatura = round(temperatura, 2)
    
    #Mostrando a temperatura na tela
    texto_temperatura = fonte.render(
    f"T* = {temperatura} u.a",
    True,
    (0,0,0)
    )

    #Mostrando na tela um pouco abaixo de N
    tela.blit(texto_temperatura, (620, 90))
```

```python
def adicionar_particulas(particulas, quantidade):
    """ Adiciona novas partículas ao sistema com posições aleatórias dentro da caixa
    e velocidades iniciais aleatórias em diferentes direções. """
    
    for _ in range(quantidade):

        if len(particulas) >= max_particulas:
            break

        angulo = rd.uniform(0, 2*mt.pi)
        velocidade = rd.uniform(2, 6)

        x = rd.randint(
            caixa_x + raio,
            caixa_x + caixa_largura - raio
        )

        y = rd.randint(
            caixa_y + raio,
            caixa_y + caixa_altura - raio
        )

        particulas.append(
            {
                "x": x,
                "y": y,
                "vx": velocidade*mt.cos(angulo),
                "vy": velocidade*mt.sin(angulo)
            }
        )
    return particulas
```

```python
def remover_particulas(particulas, quantidade):
    """ Remove uma quantidade de partículas da lista do sistema, desde que reste pelo menos uma partícula. Retorna a lista atualizada de partículas. """
    
    for _ in range(quantidade):

        if len(particulas) > min_particulas:
            particulas.pop()
        else:
            break
    return particulas
```

```python
def limitar_velocidade(particula):
    """Função que recebe uma partícula da lista de partículas e limita a sua velocidade."""
    
    vx = particula["vx"]
    vy = particula["vy"]

    velocidade = (vx**2 + vy**2)**0.5

    if velocidade > v_max:

        fator = v_max / velocidade

        particula["vx"] *= fator
        particula["vy"] *= fator

    elif velocidade < v_min:

        fator = v_min / velocidade

        particula["vx"] *= fator
        particula["vy"] *= fator
```

## Distribuição de velocidades

Para visualizar como as velocidades das partículas estão distribuídas, é construído um histograma atualizado em tempo real.

Essa representação permite comparar qualitativamente o comportamento observado com a distribuição de Maxwell-Boltzmann.

```python
def velocidades(particulas):
    """Calcula a rapidez de cada partícula da lista de partículas e retorna uma lista contendo essas velocidades."""
    lista_velocidades = []

    for particula in particulas:

        v = (particula["vx"]**2 + particula["vy"]**2)**0.5

        lista_velocidades.append(v)

    return lista_velocidades
```

```python
def histograma_velocidades(particulas):
    """Agrupa as velocidades das partículas em intervalos de velocidade e retorna uma lista com a frequência de partículas em cada intervalo."""
    velocidades_particulas = velocidades(particulas)

    barras = [0] * 10

    for v in velocidades_particulas:

        indice = int(v)

        if indice > 9:
            indice = 9

        barras[indice] += 1

    return barras
```

```python
def escala_altura(valor, max_valor):
    """Função que recebe o valor que será inserido do gráfico e corrige a sua escala de acordo com o maior valor do gráfico."""
    
    if max_valor == 0:
        return 0
    return (valor / max_valor) * max_altura
```

```python
def desenha_histograma(tela, particulas):
    """Desenha um histograma estilizado da distribuição de velocidades das partículas."""

    barras = histograma_velocidades(particulas)

    x0 = 640
    y0 = 550

    largura_barra = 18
    espacamento = 22

    max_valor = max(barras)

    texto_histograma = fonte.render(
    f"Distribuição de velocidades",
    True,
    (0,0,0)
    )

    #Mostrando na tela um pouco abaixo de N
    tela.blit(texto_histograma, (600, 300))
    
    # eixo X
    pyg.draw.line(tela, (0,0,0), (x0, y0), (x0 + 250, y0), 2)

    # eixo Y
    pyg.draw.line(tela, (0,0,0), (x0, y0), (x0, y0 - max_altura), 2)

    for i, valor in enumerate(barras):

        altura = escala_altura(valor, max_valor)

        # cor suave estilo gráfico científico
        cor = (50, 120, 200)

        pyg.draw.rect(
            tela,
            cor,
            (
                x0 + i * espacamento,
                y0 - altura,
                largura_barra,
                altura
            )
        )
    
    #rótulo do eixo x
    for i in range(10):
        texto = fonte.render(str(i), True, (0,0,0))
        tela.blit(texto, (x0 + i*22, y0 + 5))

        passos = 5  # número de divisões no eixo Y

    for i in range(passos + 1):

        valor = int((max_valor / passos) * i)
        y = y0 - (max_altura / passos) * i
    
        texto = fonte.render(str(valor), True, (0,0,0))
    
        # desloca para esquerda do eixo Y
        tela.blit(texto, (x0 - 40, y - 10))
    
        # linhas guias (deixa MUITO mais bonito)
        pyg.draw.line(
            tela,
            (220, 220, 220),
            (x0, y),
            (x0 + 250, y),
            1
        )
```

## Execução da simulação

Nesta etapa são definidos os parâmetros iniciais do sistema e iniciado o loop principal responsável pela atualização contínua da interface gráfica.

```python
pyg.init() #inicializa o pygame

#Cria a janela e define o seu tamanho
tela = pyg.display.set_mode((900,620)) 

pyg.display.set_caption("Ideal Gas Lab")

#loop principal
rodando = True

clock = pyg.time.Clock()

min_particulas = 1
max_particulas = 300


#visual do grafico e altura máxima visual do gráfico
max_barras = 10
max_altura = 200 

#velocidade maxima e minima para as particulas
v_max = 8.0
v_min = 0.5

#Criação do recipiente do gás
caixa_x = 25
caixa_y = 25
caixa_largura = 550
caixa_altura = 550

#estabelecendo um padrão para posicinar as mensagens das teclas que alteram a temperatura e o número de partículas
painel_x = 620
painel_y = 140
espaco_linha = 35

#quantidade inicial e o raio das partículas
raio = 8
num_particulas = 100

#Definindo a fonte que mostrará a mensagem na tela.
fonte = pyg.font.Font(None, 30)

#lista que armazenará as partículas
particulas = []

cria_particulas(particulas, num_particulas)


#loop em que a simulação ocorrerá
while rodando:
    

    #captura eventos
    rodando = trata_evento(particulas, rodando)
            
    #fundo da tela cinza
    tela.fill((230, 230, 230))

    texto_particulas(tela, fonte, particulas, max_particulas)

    mover_particulas(particulas)

    desenha_caixa(tela, caixa_x, caixa_y, caixa_largura, caixa_altura)
    
    desenha_particulas(particulas, raio, tela)
        
    colisoes_parede(particulas)

    colisoes_particulas(particulas, raio)

    temp(particulas, fonte, tela)
    
    controles(fonte, tela, painel_x, painel_y)

    desenha_histograma(tela, particulas)

    #(60 fps)
    clock.tick(60)
    
    #atualização da tela
    pyg.display.flip()




pyg.quit()
```
